```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'factor': ['basic_eps', 'diluted_eps', 'pb', 'pe', 'ps', 'circ_mv', 'total_mv_log', 
               'tangibleasset_to_deb', 'n_income', 'assets_turn', 'macd', 'wr10', 'rsi', 
               'bias', 'cci', 'kdj_j', 'ar', 'br', 'sar', 'obv'],
    'Ic': [5.67, 3.22, -2.07, -2.11, -2.14, -2.13, -3.13, 2.83, 2.01, 2.23, -3.10, 2.84, 
           -2.01, -2.42, -2.48, -2.84, -2.01, -2.04, -3.21, -3.81],
    'Ir': [0.33, 0.33, -0.19, -0.21, -0.21, -0.17, -0.23, 0.38, 0.27, 0.25, -0.31, 0.26, 
           -0.18, -0.21, -0.23, -0.26, -0.22, -0.22, -0.31, -0.36]
}

df = pd.DataFrame(data)

correlation_matrix = df.drop(columns=['factor']).corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar=True)
plt.title('Heatmap of Correlation')
plt.xticks(np.arange(0.5, len(correlation_matrix.columns), 1), correlation_matrix.columns)
plt.yticks(np.arange(0.5, len(correlation_matrix.index), 1), correlation_matrix.index)
plt.tight_layout()

heatmap_file_path = 'heatmap.png'
plt.savefig(heatmap_file_path)

heatmap_file_path
```
